```python
import pandas as pd

# Clean up any whitespace and make sure both are strings
df_services['customer_service_customer_id'] = df_services['customer_service_customer_id'].astype(str).str.strip()
df_customers['customer_id'] = df_customers['customer_id'].astype(str).str.strip()

# 1. Calculate matching statistics
services_ids = set(df_services['customer_service_customer_id'].unique())
customer_ids = set(df_customers['customer_id'].unique())

matched_ids = services_ids.intersection(customer_ids)
missing_in_customer = services_ids - customer_ids

print("=== COUPLING VERIFICATION STATS ===")
print(f"Total unique customers in Services table: {len(services_ids)}")
print(f"Total unique customers in Customer master table: {len(customer_ids)}")
print(f"IDs that match perfectly between both tables: {len(matched_ids)}")
print(f"IDs in Services that are missing from Customer table: {len(missing_in_customer)}")
print("="*45 + "\n")

# 2. Print a side-by-side visual sample to check the names
print("=== VISUAL COMPARISON SAMPLE ===")
# Grab 3 random rows where an ID exists in both datasets
sample_ids = list(matched_ids)[:3]

for cid in sample_ids:
    # Extract names using the ID
    service_name = df_services[df_services['customer_service_customer_id'] == cid]['customer_service_customer_name'].values[0]
    customer_table_name = df_customers[df_customers['customer_id'] == cid]['customer_name'].values[0]
    
    print(f"Checking ID: {cid}")
    print(f"  -> Name in Services Table: '{service_name}'")
    print(f"  -> Name in Customer Table: '{customer_table_name}'")
    print("-" * 40)
```

```python
import pandas as pd

# Clean up company name strings to ensure minor spaces don't ruin the match
df_services['clean_service_name'] = df_services['customer_service_customer_name'].astype(str).str.strip().str.lower()
df_customers['clean_customer_name'] = df_customers['customer_name'].astype(str).str.strip().str.lower()

# Calculate matching stats based on names
service_names = set(df_services['clean_service_name'].unique())
customer_names = set(df_customers['clean_customer_name'].unique())

matched_names = service_names.intersection(customer_names)

print("=== NAME-BASED COUPLING VERIFICATION ===")
print(f"Total unique customer names in Services: {len(service_names)}")
print(f"Total unique customer names in Customer Table: {len(customer_names)}")
print(f"Names that match PERFECTLY: {len(matched_names)}")
print(f"Match Rate: {(len(matched_names) / len(service_names)) * 100:.1f}%")
print("="*45 + "\n")

# Let's peek at a couple of successful name matches to inspect their underlying IDs
if len(matched_names) > 0:
    sample_name = list(matched_names)[0]
    s_id = df_services[df_services['clean_service_name'] == sample_name]['customer_service_customer_id'].values[0]
    c_id = df_customers[df_customers['clean_customer_name'] == sample_name]['customer_id'].values[0]
    
    # Check if there is another column in df_customers holding the service ID
    c_row = df_customers[df_customers['clean_customer_name'] == sample_name].iloc[0]
    
    print(f"Visual Verification for match: '{sample_name}'")
    print(f"  -> ID in Services Table: {s_id}")
    print(f"  -> ID in Customer Table: {c_id}")
    print("\n--- Let's look for where that Service ID might be hiding in the Customer table ---")
    for col in df_customers.columns:
        if str(c_row[col]) == str(s_id):
            print(f"💡 FOUND A MATCH! The customer column '{col}' contains the value {s_id}!")
```
